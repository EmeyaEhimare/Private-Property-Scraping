```python
rentals = {
    'title':[],
    'location':[],
    'description':[]
    
}
```

```python
for i in range(0,105):
    url = f'https://www.privateproperty.com.ng/property-for-rent?page={i}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    page = soup.find(class_ = 'result-listings')
    apartments = page.find_all(class_ = 'similar-listings-info')
   
    
    

    for apartment in apartments:
        description = apartment.find('a').text
        location = apartment.find('p', class_ = 'listings-location').text.strip()
        title = apartment.find('h3').text


        rentals['title'].append(title)
        rentals['location'].append(location)
        rentals['description'].append(description)



           #location = apartment.find('p', class_ = 'listings-location')
           #title_link = apartment.find('a')
           #title = title_link.get_text()
           #price = apartment.find(class_ ='col-md-4')

```

```python
rent_df = pd.DataFrame(rentals)
```

```python
prices = {
    'price' : [],
    'number of bedrooms' : [],
    'number of bathrooms': [],
    'number of toilets' : []
}
```

```python
for i in range(0,105):
    url = f'https://www.privateproperty.com.ng/property-for-rent?page={i}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    pricing = soup.find_all(class_ = 'similar-listings-price')
    
    for price in pricing:
        rent_price = price.find('span').find_next('span').text
        if '/year' in rent_price:
            rent_price = rent_price.replace("/year", "")
        number_of_bedrooms = price.find('li').text.strip()
        number_of_bathrooms = price.find('li').find_next('li').text.strip()
        number_of_toilets = price.find('li').find_next('li').find_next('li').text.strip()
        #rent_price.replace('/year','')
        #print(rent_price)

        prices['price'].append(rent_price)
        prices['number of bathrooms'].append(number_of_bathrooms)
        prices['number of bedrooms'].append(number_of_bedrooms)
        prices['number of toilets'].append(number_of_toilets)
    
```

```python
rent_price = pd.DataFrame(prices)
```

```python
rent_price = rent_price.rename(columns={'price':'price per year'})
```

```python
df =pd.concat([rent_df,rent_price], axis = 1,  ignore_index = False)
```
